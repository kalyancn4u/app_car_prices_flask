## 0 · The big picture, in plain English

Four ideas, with everyday analogies:

| Term | What it really is | Analogy |
| :--- | :---------------- | :------ |
| **Docker image** | A frozen snapshot of your app **plus everything it needs** (Python, libraries, the model files). | A **recipe + all the ingredients** sealed in a box. Nothing is "missing on the other computer". |
| **Docker container** | A **running copy** of an image. | The **cake baked** from the recipe. You can bake many identical cakes from one recipe. |
| **AWS ECR** | A **private online locker** for your images. | **Google Photos, but for Docker images.** You upload your image here so AWS can find it. |
| **AWS ECS + Fargate** | The service that **runs** your container in the cloud. **Fargate** means *you don't rent or babysit any servers* — AWS runs it and bills per use. | A **taxi**: you don't own or maintain the car, you just say "drive this" and pay for the trip. |

**The whole journey is just this:**

```
 your laptop                          AWS cloud
 ┌───────────┐   docker push   ┌────────┐   runs on    ┌──────────────┐
 │  Docker   │ ───────────────►│  ECR   │ ────────────►│ ECS Fargate  │──► public IP:5000
 │  image    │                 │(locker)│              │ (the taxi)   │     /predict
 └───────────┘                 └────────┘              └──────────────┘
```

> 🔑 **The golden rule of Docker:** *if it runs in the container on your laptop, it runs
> identically on AWS* — because the container carries its whole world with it. That's why
> we test locally first (Parts 2–3) before touching the cloud.

## 3½ · ✈️ Pre-flight: point the AWS CLI at the **right** account

You have **one username/password per Scaler exam, plus your own AWS account, all on the same
laptop.** That's exactly the setup where it's easy to run a command against the **wrong**
account. So before a single `aws` command, do a 10-second identity check.

### 🔑 First, a crucial distinction (trips up almost everyone)

| | The **Console** (website) | The **AWS CLI** (terminal) |
| :-- | :-- | :-- |
| Logs in with… | **username + password** (+ MFA) | **Access keys** — an *Access Key ID* + *Secret Access Key* |
| It **cannot**… | run `docker push` / `aws ecs …` by itself | be logged in with a username/password |

So a username/password **by itself does nothing for the CLI**. For the CLI you need that
account's **access keys**. Two ways to get unstuck:

- **If you can create keys:** Console → **IAM → Users → (your user) → Security credentials →
  Create access key → "Command Line Interface (CLI)"** → copy the *Access Key ID* and the
  *Secret Access Key* (the secret is shown **only once**).
- **If the exam gives you only a Console login (no keys):** use **AWS CloudShell** — click the
  **`>_`** icon in the Console's top bar. It's a terminal **already signed in as your Console
  user**, so there are *no keys to configure*. 🎓 Ideal for exams — just run every `aws`
  command there and skip the profile setup below.

### 1️⃣ "Who am I right now?" — always check first

```bash
aws sts get-caller-identity
```
Example output:
```json
{
  "UserId":  "AIDA................",
  "Account": "123456789012",     <-- does this 12-digit ID match your EXAM account?
  "Arn":     "arn:aws:iam::123456789012:user/exam-user"
}
```
👉 **Match the `Account` number to the exam you're actually doing** before anything else.

### 2️⃣ Which profiles exist, and which one is active?

Because you juggle several accounts, give **each its own named profile** instead of
overwriting one set of keys:

```bash
aws configure list-profiles      # every profile you've set up
aws configure list               # the ACTIVE profile: keys (masked), region, and their source
```

### 3️⃣ Set up / update a profile for a specific account

```bash
aws configure --profile scaler-exam
#   AWS Access Key ID     : <that account's key id>
#   AWS Secret Access Key : <that account's secret>
#   Default region name   : us-east-1
#   Default output format : json
```
Do this once per account (e.g. `scaler-exam`, `personal`). Credentials are saved under
`~/.aws/credentials` (Windows: `C:\Users\<you>\.aws\credentials`) — one `[section]` per profile.

### 4️⃣ Switch to the right one — two ways

**Per command** (explicit and safest — hard to get wrong):
```bash
aws sts get-caller-identity --profile scaler-exam
# and add --profile scaler-exam to the ECR/ECS commands in Parts 4-7 too
```

**For the whole terminal session** (set it once, then every `aws` uses it):
```bash
# macOS / Linux
export AWS_PROFILE=scaler-exam
# Windows PowerShell
$env:AWS_PROFILE = "scaler-exam"
# Windows CMD
set AWS_PROFILE=scaler-exam
```
Then re-run `aws sts get-caller-identity` to **confirm it flipped** to the right account.

### 5️⃣ Region check (mixing Regions is the #1 "it says not found" bug)

```bash
aws configure get region                                   # which Region is active?
aws configure set region us-east-1 --profile scaler-exam   # set it if it's wrong
```

### ✅ The 10-second pre-flight ritual — before **every** AWS session
1. `aws sts get-caller-identity` → does **Account** match the exam account? ✅
2. `aws configure get region` → is it the Region you intend? ✅
3. *Only now* run the ECR / ECS commands in Parts 4-6 — either add `--profile scaler-exam`
   to each command, or set `AWS_PROFILE` once (step 4).

> 🧭 **The habit:** *"Check who I am, check where I am, **then** act."* One wrong-account
> `docker push` or `ecs run-task` is how exam resources land in your personal bill — this
> 10-second check prevents it.

## 4 · Push the image to **AWS ECR** (the cloud locker)

ECR is a private place to store your image so ECS can pull it. The Console makes this
delightfully easy — **it writes the exact commands for you.**

### 🖱️ In the AWS Console (click-by-click)
1. Sign in at **[console.aws.amazon.com](https://console.aws.amazon.com)** with your username/password.
2. Check the **Region** (top-right) is the one you'll use, e.g. **N. Virginia (us-east-1)**.
3. In the top search bar type **ECR** → open **Elastic Container Registry**.
4. **Create repository** → **Private** → name it exactly **`car-price-api`** → **Create**.
5. Open the repo → click **View push commands** (top-right button).
6. AWS shows **4 ready-made commands** tailored to your account. Run them in a terminal
   **from this repo's folder**. They do: (1) log Docker in to ECR, (2) build, (3) tag, (4) push.

### ⌨️ The 4 commands (what "View push commands" gives you)
Replace `<ACCOUNT_ID>` and `<REGION>` with yours (the Console fills these in automatically):

```bash
# 1) Log Docker in to your ECR
aws ecr get-login-password --region <REGION> | \
  docker login --username AWS --password-stdin <ACCOUNT_ID>.dkr.ecr.<REGION>.amazonaws.com

# 2) Build (you already did this)
docker build -t car-price-api .

# 3) Tag the image with its full ECR address
docker tag car-price-api:latest <ACCOUNT_ID>.dkr.ecr.<REGION>.amazonaws.com/car-price-api:latest

# 4) Push it up to the cloud
docker push <ACCOUNT_ID>.dkr.ecr.<REGION>.amazonaws.com/car-price-api:latest
```

> 🍎 **Apple-Silicon / M1–M4 Mac users:** Fargate runs **amd64**. Build with
> `docker build --platform linux/amd64 -t car-price-api .` (step 2 above) or the task will
> crash-loop on AWS. On Windows/Intel this is already the default.

When it finishes, refresh the ECR repo page — you'll see your image with tag `latest`. 🎉

## 5 · Run it on **ECS Fargate** (the taxi that runs your container)

Three nested ideas, simplest first:
- **Task definition** = the *blueprint*: "run **this image**, give it this much CPU/RAM, open port 5000".
- **Task** = one *running instance* of that blueprint.
- **Cluster** = the *logical group* the tasks live in.
- **Fargate** = the *launch type* meaning "AWS supplies the compute; I manage nothing".

### 🖱️ In the AWS Console (click-by-click)

**A. Create a cluster**
1. Search **ECS** → **Elastic Container Service** → **Clusters** → **Create cluster**.
2. Name: **`car-price-cluster`**. Under *Infrastructure* leave **AWS Fargate (serverless)** ticked → **Create**.

**B. Create a task definition**
1. ECS → **Task definitions** → **Create new task definition**.
2. Family name: **`car-price-api`**. Launch type: **AWS Fargate**.
3. CPU **0.5 vCPU**, Memory **1 GB** (plenty for this app).
4. **Container details:**
   - Name: **`car-price-api`**
   - **Image URI:** paste your ECR address → `<ACCOUNT_ID>.dkr.ecr.<REGION>.amazonaws.com/car-price-api:latest`
   - **Port mappings:** container port **5000**, protocol **TCP**.
5. (Recommended) turn **on** *Use log collection* (CloudWatch) so you can read logs.
6. **Create.**

**C. Run the task**
1. Open your **cluster** → **Tasks** tab → **Run new task** (or create a **Service** to keep it always-on).
2. Launch type **Fargate**; Task definition **`car-price-api`** (latest revision).
3. **Networking:**
   - Pick your **VPC** and a **public subnet**.
   - **Auto-assign public IP: ENABLED** ← *essential*, or you can't reach it.
   - **Security group → Create new →** add an **inbound rule**: type **Custom TCP**, port **5000**,
     source **Anywhere (0.0.0.0/0)** *(fine for a short test; restrict to your IP for anything real)*.
4. **Run task.** Wait ~1 minute until **Last status = RUNNING**.

**D. Find the public address**
1. Click the running **task** → **Networking** section → copy the **Public IP**.
2. Your API is now live at **`http://<PUBLIC_IP>:5000`**.

> 🧠 **What just happened:** ECS pulled your image from ECR and started the container on
> hardware AWS owns. You never launched a server — that's the "Fargate / serverless" magic.

## 7 · 🧹 Complete cleanup — leave **nothing** behind (avoid surprise bills)

When you deploy, AWS quietly creates **several linked resources**. If you delete only the
obvious one (the task) and forget the rest, some keep **costing money** (a running Fargate
task, stored logs, stored images) or just clutter your account. This section removes
**every** piece, **in the right order**, and then **verifies** nothing is left.

### First: what you created, and what actually costs money

| Resource | Created when… | Costs money? | Delete it? |
| :------- | :------------ | :----------- | :--------- |
| **ECS task** (Fargate) | you *Ran the task* | 💰 **YES** — billed *per second while RUNNING* | ✅ stop it |
| **ECS service** (if you made one) | optional | 💰 **YES** — it keeps re-launching tasks | ✅ delete it |
| **ECS cluster** | you created it | Free (when empty) | ✅ tidy up |
| **Task definition** | you registered it | Free | ⬜ optional (deregister) |
| **CloudWatch log group** `/ecs/car-price-api` | logging was on | 💰 small (log storage) | ✅ delete it |
| **ECR repo + image** | you pushed | 💰 small (image storage) | ✅ delete it |
| **Security group** (custom, port 5000) | you created it | Free | ✅ tidy up (after task gone) |
| **IAM role** `ecsTaskExecutionRole` | first ECS use | Free, **reusable** | ⬜ **leave it** (shared) |
| **Default VPC / subnets** | already existed | Free | ❌ **NEVER delete** |

> 🔑 **The short version:** what *costs money if left running* is the **Fargate
> task/service**. What *costs a little if left stored* is the **CloudWatch log group** and
> the **ECR image**. Kill those three and you're safe — the rest is just tidiness.

### Teardown order (dependencies matter — do these top-to-bottom)

Everything else counts as "in use" until the **task is stopped**, so that goes first.

**Step 1 — Stop the task / delete the service**
- 🖱️ **Console:** ECS → your cluster.
  - If you created a **Service:** open it → **Update** → *Desired tasks* = **0** → **Update**, then **Delete** the service.
  - If you only **Ran a task:** **Tasks** tab → tick the task → **Stop**.
- ⌨️ **CLI:**
  ```bash
  # if a service:
  aws ecs update-service --cluster car-price-cluster --service <SERVICE_NAME> --desired-count 0 --region <REGION>
  aws ecs delete-service  --cluster car-price-cluster --service <SERVICE_NAME> --force --region <REGION>
  # if a standalone task:
  aws ecs stop-task --cluster car-price-cluster --task <TASK_ARN> --region <REGION>
  ```

**Step 2 — Delete the cluster**
- 🖱️ ECS → **Clusters** → `car-price-cluster` → **Delete cluster**.
- ⌨️ `aws ecs delete-cluster --cluster car-price-cluster --region <REGION>`

**Step 3 — Deregister the task definition(s)** *(free, but keeps the list clean)*
- 🖱️ ECS → **Task definitions** → `car-price-api` → tick each revision → **Actions → Deregister**.
- ⌨️ `aws ecs deregister-task-definition --task-definition car-price-api:1 --region <REGION>`

**Step 4 — Delete the CloudWatch log group** *(easy to forget — it silently keeps storing logs)*
- 🖱️ **CloudWatch** → **Log groups** → search `/ecs/car-price-api` → tick it → **Actions → Delete**.
- ⌨️ `aws logs delete-log-group --log-group-name /ecs/car-price-api --region <REGION>`

**Step 5 — Delete the ECR image + repository**
- 🖱️ **ECR** → `car-price-api` → tick the image(s) → **Delete**; then the repo → **Delete**.
- ⌨️ `aws ecr delete-repository --repository-name car-price-api --force --region <REGION>`

**Step 6 — Delete the custom security group** *(only works once the task has fully stopped)*
- 🖱️ **EC2** → **Security Groups** → find the one you made (inbound **TCP 5000**) → **Actions → Delete security groups**.
  *(You cannot delete the `default` group — that's normal, leave it.)*
- ⌨️ `aws ec2 delete-security-group --group-id <SG_ID> --region <REGION>`
- ⏳ If it says **"in use"**, the task's network interface hasn't released yet — wait ~1–2 min and retry.

**Step 7 — Load balancer / target group / Elastic IP** *(only if you added them)*
- The basic Part 5 flow **does not** create these, so you most likely have none. **But** if you
  attached an **Application Load Balancer**, delete the **load balancer** *and* its **target
  group** (ALBs bill *by the hour*, even idle). If you manually **allocated an Elastic IP**,
  **release** it. Fargate's auto-assigned public IP frees itself when the task stops — nothing to do.

**Step 8 — Local (your laptop) — optional**
```bash
docker rm -f car-price-api                     # remove the container if still running
docker image rm car-price-api car-price-api:dev  # reclaim ~1.5 GB of disk
```

### ✅ Verify it's all gone (the 60-second sweep)

Open each page and confirm it's empty — *seeing it with your own eyes* is the point:

- [ ] **ECS → Clusters** → no `car-price-cluster` (or zero running tasks/services)
- [ ] **ECS → Task definitions** → `car-price-api` deregistered (greyed) or absent
- [ ] **ECR → Repositories** → no `car-price-api`
- [ ] **CloudWatch → Log groups** → no `/ecs/car-price-api`
- [ ] **EC2 → Security Groups** → your custom port-5000 group is gone
- [ ] **EC2 → Load Balancers** and **Elastic IPs** → nothing you created remains

### 🛡️ Safety net — set a billing alarm once (highly recommended)

Even after a clean teardown, this emails you if anything *ever* costs more than a dollar:

- 🖱️ **Billing and Cost Management** → **Budgets** → **Create budget** → **Zero spend budget**
  (or a **$1 monthly cost budget**) → enter your **email** → **Create**.
- To see current charges: **Billing → Bills**, or **Cost Explorer** (figures update within ~24h).

> 💡 **The one-sentence habit:** *stop the task, delete the cluster, the ECR repo, and the
> `/ecs/…` log group; then eyeball each service page until it's empty — and keep a $1 billing
> alarm.* Do that every time and the free tier stays free. 🟢

## 8 · Troubleshooting (the usual suspects)

| Symptom | Cause & fix |
| :------ | :---------- |
| `curl: (7) Failed to connect` to the public IP | Security group doesn't allow **inbound TCP 5000**, or **public IP wasn't enabled** on the task. Re-check step 5C. |
| Task shows **STOPPED** seconds after starting | Wrong CPU architecture — rebuild with `docker build --platform linux/amd64 …`, re-push, run again. Check the task's **Logs** tab. |
| `405 Method Not Allowed` in a browser | **Expected.** Browsers send `GET`; `/predict` needs `POST`. Use `curl -X POST` or Postman. |
| `{"status":"unhealthy"}` / `503` | The `models/` folder wasn't in the image. Confirm `COPY models/ ./models/` is in the `Dockerfile` and the folder existed **before** you built. |
| `400 Unknown make/model` | Typo, or a car not in the training data. Casing doesn't matter (it's uppercased), spelling does. |
| ECS can't pull the image | The task definition's **Image URI** is wrong, or ECR is in a **different Region** than the cluster. Keep one Region throughout. |
| Security group **"in use"** when deleting | The task/network-interface hasn't released yet. Wait 1–2 min after the task stops, then retry. |
